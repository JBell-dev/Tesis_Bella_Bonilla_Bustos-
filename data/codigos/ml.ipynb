```python
#aca meto los graphers
def binaryGrapher(x,y,df,line='no'):  
  #print('--------------------------------------------df--------------------------')
  #print(df,'este es df')
  a=y[0]
  b=y[1]
  #print('a es',a)
  #print('b es',b)
  aux1=df[df.columns.values]
  aux2=df[df.columns.values]
  
  df0=aux1
  df0['color']=a
  df0['why']=aux1.loc[:,a]

  df1=aux2

  df1['color']=y[1]
  df1['why']=aux2.loc[:,y[1]]
  #print('--------------------------------------------auxiliares--------------------------')

  #print(df0,df1,'estos son los dfs')
  result = df0.append(df1)
  #print('--------------------------------------------resultado--------------------------')

  #print(result)
  if line=='yes':
    px.line(result,x=x,y='why',color='color').show()
  else:
    px.scatter(result,x=x,y='why',color='color').show()

def predictGraph(predictDf,name):
  
  #primero genero el que voy a concatenar abajo
  col=predictDf.columns
  auxDf=predictDf[['Borough','Year',name]]
  q=len(predictDf.columns)-3
  for i in np.arange(0,q,1):
    auxDf[col[i+3]]=auxDf[name]

  auxDf['what']='real'
  predictDf['what']='predict'
  result=pd.concat([auxDf,predictDf])

  return result

```
