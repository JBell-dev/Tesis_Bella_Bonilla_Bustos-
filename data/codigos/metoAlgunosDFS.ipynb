```python
def rename_unnamed(df):
    """Rename unamed columns name for Pandas DataFrame

    See https://stackoverflow.com/questions/41221079/rename-multiindex-columns-in-pandas

    Parameters
    ----------
    df : pd.DataFrame object
        Input dataframe

    Returns
    -------
    pd.DataFrame
        Output dataframe

    """
    for i, columns in enumerate(df.columns.levels):
        columns_new = columns.tolist()
        for j, row in enumerate(columns_new):
            if "Unnamed: " in row:
                columns_new[j] = ""
        if pd.__version__ < "0.21.0":  # https://stackoverflow.com/a/48186976/716469
            df.columns.set_levels(columns_new, level=i, inplace=True)
        else:
            df = df.rename(columns=dict(zip(columns.tolist(), columns_new)),
                           level=i)
    return df
def first(year):
  df0=pd.read_excel('/content/drive/MyDrive/Tesis Economia/Cambio climatico/agus/pre/cob-borough.xls',sheet_name=year,header=[0, 1,2,3])
  df0=df0.rename(columns={'Area Code':'Code','Area Name':'Borough'})
  df0=df0.drop(np.arange(33,54,1))
  df0=df0.drop('CI +/-',axis=1,level=3)
  df1=df0[['Code','Borough']]
  df2=df0.drop(['Code','Borough'],axis=1)
  df2 = df2.apply(pd.to_numeric, args=('coerce',))
  df3=pd.concat([df1,df2],axis=1)
  df3['Year']=int(year)
  return df3
dfv=first('2004')
acum=dfv.append(first('2005'),ignore_index=True)
for i in np.arange(2006,2019,1):
  dfv=first(str(i))
  acum=acum.append(dfv)
acum.reset_index(inplace=True)
acum=acum.drop('index',axis=1,level=0)
#acum1=acum.set_index(['Borough','Year'],inplace=True)
acum=rename_unnamed(acum)
acum=acum.droplevel(3, axis=1)

#acum.loc[acum['Borough']=='Barnet']
#print(acum.columns.value_counts())
#(Non-European Union, Asia, South East Asia)  
acum = acum.loc[:,~acum.columns.duplicated()]
acum.set_index(['Borough','Year'],inplace=True)
acum.sort_index(inplace=True)
#Barking and Dagenham
#acum.loc[(slice(None),2004),:]
#acum.to_csv('cob-borough.csv')
#acum.to_pickle('cob-borough.pkl')
acum
```
